# f) Training & predicting the model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
pred_slr = lr.predict(X_test)
```

# c) Splitting the dataset into train & test set

```python
x_mlr = df.drop(columns=['quality'])
y_mlr = df['quality']
X_train_mlr, X_test_mlr, y_train_mlr, y_test_mlr = train_test_split(x_mlr, y_mlr, random_state = 13, test_size = 0.2)
```

# d)  Training the model & predicting

```python
mlr = LinearRegression()
mlr.fit(X_train_mlr, y_train_mlr)
```

```python
pred_mlr = mlr.predict(X_test_mlr)
```

# 3. Diagnostic Plots

# i) SLR

# a) Checking homoscedasticity of residues

```python
res_lr = y_test - pred_slr
```

```python
plt.title("Residual plot of SLR", size = 16)
sns.scatterplot(pred_slr, res_lr)
```

# b) Checking Normality of residues

```python
import scipy as sp

plt.figure(figsize=(6,4))
sns.kdeplot(res_lr)
plt.title("PDF plot of residuals in SLR", size = 16)

fig, ax = plt.subplots(figsize=(6,4))
sp.stats.probplot(res_lr, plot = ax, fit=True)
plt.title("Q-Q plot for SLR", size = 16)
plt.show()
```

# 4. RESULTS - Interpreting the dataset 

# ii) Performance of models

# SLR Model

```python
from sklearn.metrics import r2_score, mean_absolute_error

r2_score_slr = round(r2_score(y_test, pred_slr),3)
mean_slr = round(mean_absolute_error(y_test, pred_slr),3)
adj_r2_slr = round(1-((1-r2_score_slr)*(793-1)/(793-1-1)),3)

print("R2 score = ", r2_score_slr)
print("Adjusted R2 score = ", adj_r2_slr)
print("Mean Absolute Error = ", mean_slr)
```

# MLR Model

```python
r2_score_mlr = round(r2_score(y_test_mlr, pred_mlr),3)
mean_mlr = round(mean_absolute_error(y_test_mlr, pred_mlr),3)
adj_r2_mlr = round(1-((1-r2_score_mlr)*(793-1)/(793-1-1)),3)

print("R2 score = ", r2_score_mlr)
print("Adjusted R2 score = ", adj_r2_mlr)
print("Mean Absolute Error = ", mean_mlr)
```
